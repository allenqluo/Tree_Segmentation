```python
import matplotlib.ticker as plticker
from sklearn.linear_model import LinearRegression
from matplotlib.colors import to_rgba

selected_domains = ['D01', 'D02', 'D03', 'D04',
                    'D05', 'D06', 'D07', 'D08',
                    'D10', 'D11', 'D12', 'D13',
                    'D15', 'D16', 'D17', 'D19', 'D20']


## Plotting and output params
outdir = "../figs/veg-struct/allom/"
# Choose structural variable to plot

xvars_list = ["stemDiameter", "stemDiameter","height","height"]
yvars_list = ["height", "crownRadius","crownRadius","crown_vol"]

xlabs_list = ["ln(D)", "ln(D)","ln(H)", "ln(H)"]
ylabs_list = ["ln(H)", r"ln($R_{hc}$)",r"ln($R_{hc}$)", r"ln($V_{c}$)"]

for var_idx, yvar in enumerate(yvars_list):
    
    # get xvar and labels
    xvar = xvars_list[var_idx]
    xlabel = xlabs_list[var_idx]
    ylabel = ylabs_list[var_idx]
    
    
    # Create figure
    fig, axs = plt.subplots(6,3, figsize=(8,12), sharex=True, sharey=True)
    ax = axs.flatten()

    # Loop through domains and plot allom
    plot_idx = 0
    for domain, group in appidv_all_clean.groupby("domainID"):

        if domain not in selected_domains:
            # use selected domains only! (with crownvol measurements/crown diameter)
            continue

        ## H ~ D (natural log)
        log_x = np.log(group[xvar].values)
        log_y = np.log(group[yvar].values)
        ax[plot_idx].scatter(log_x,log_y,s=18,
                             color=to_rgba(mpl_colors[var_idx], alpha=.7),
                             ec=to_rgba("#141414", alpha=.95),
                             lw=0.3)

        # Add regression
        boolfilt = np.isfinite(log_x) & np.isfinite(log_y) # filter out nan and inf
        if len(log_x[boolfilt])==0:
            # not enough samples in filtered dataset
            print("Not enough samples in ", domain)
            # plot_idx+=1
            continue
        lr_model = LinearRegression(fit_intercept=True).fit(log_x[boolfilt].reshape(-1,1),
                                                            log_y[boolfilt].reshape(-1,1))
        slope, intercept = lr_model.coef_[0], lr_model.intercept_
        rsquared = lr_model.score(log_x[boolfilt].reshape(-1,1),
                                  log_y[boolfilt].reshape(-1,1))
        ypred = lr_model.predict(log_x[boolfilt].reshape(-1,1))
        ax[plot_idx].plot(log_x[boolfilt], ypred, color="black",lw=0.9)
        if intercept<0:
            oper = "-"
        else:
            oper="+"

        # Add regression params and eval metrics as text
        rmse = mean_squared_error(y_true=log_x[boolfilt].reshape(-1,1),
                                  y_pred=log_y[boolfilt].reshape(-1,1), squared=False)
        txt = f"y={np.round(slope,2)[0]}*X {oper} {np.absolute(np.round(intercept,2))[0]}\n" \
        f"RMSE={np.round(rmse,2)}\n"+ "$R^{2}=$" + f"{np.round(rsquared,2)}"
        # Add text
        ax[plot_idx].text(.01,.975, txt,
                          transform=ax[plot_idx].transAxes, ha="left",va="top",
                          fontweight=400,fontsize=14)
        ax[plot_idx].text(.975,.01, f"n={len(log_y[boolfilt])}",
                          transform=ax[plot_idx].transAxes, ha="right",va="bottom",
                          fontweight=400,fontsize=14)
        # add ticks to subplots
    #     ax[plot_idx].xaxis.set_major_locator(loc)
        ax[plot_idx].xaxis.set_major_locator(plticker.MaxNLocator(4))
        ax[plot_idx].yaxis.set_major_locator(plticker.MaxNLocator(5))

        # figure aesthetics
        ax[plot_idx].set_title(domain,fontsize=18,fontweight="bold")
        ax[plot_idx].set_xlabel(None)
        ax[plot_idx].set_ylabel(None)
        
        if yvar == "crownRadius":
            ax[plot_idx].set_ylim(-3,3.2)
        elif yvar=="crown_vol":
            ax[plot_idx].set_ylim(-6,10)

        # print(domain)
        # increment plot idx
        plot_idx+=1
        
        
#     if yvar == "crownRadius":
#         plt.setp(ax, xlim=-1, ylim=2)
# #         ax[plot_idx].set_ylim(-1,2)
#     elif yvar=="crown_vol":
#         ax[plot_idx].set_ylim(0,10)
    

    # Figure aesthetics
    fig.text(0.005, 0.5, ylabel, va='center', rotation='vertical')
    fig.text(0.5, 0.005, xlabel, ha='center')

    # Remove last empty axes
    ax[-1].remove()
    fig.tight_layout()
    fig.subplots_adjust(bottom=0.05,left=0.1)
    # save to output folder
    fig.savefig(os.path.join(outdir, f"DOMAINS-allom-{yvar}-{xvar}.png"),dpi=300)
#     plt.show()
```

```python

def plot_data(x,y,ax, plot_idx=0, label=None, alpha=None, beta=None, c = None, ec = None):
    
    # Add regression params
    # vstack xy for easier boolean
    filt_bool = (np.isfinite(x) & np.isfinite(y))
    xy = np.vstack([x,y]).T
    xy = xy[filt_bool]
    
    ## Linear regression
    # Get predicted y
    lr_model = LinearRegression(fit_intercept=True).fit(xy[:,0].reshape(-1,1), xy[:,1])
    slope, rsquared = lr_model.coef_[0], lr_model.score(xy[:,0].reshape(-1, 1),
                                                  xy[:,1])
    y_pred = slope*xy[:, 0] + lr_model.intercept_
    slope_ols = slope # to return original intercept
    if lr_model.intercept_<0:
        oper="-"
    else:
        oper="+"
    
    # Get colors
    c = to_rgba(c, 0.7)
    ec = to_rgba(ec, 1)
    
    # Plot
    ax.scatter(xy[:,0],xy[:,1],
        label=label,
        c = c,
        edgecolors = ec)
    ax.plot(xy[:,0], y_pred, color="#414142", alpha=0.85)

    # Plot with color density
#     # Get gaussian density from data to plot density in color
#     values = np.vstack([xy[:,0], xy[:,1]])
#     kernel = gaussian_kde(values)(values)
#     ax.scatter(xy[:,0],xy[:,1],
#                label=label,
#                c=kernel,vmin=np.nanmin(kernel)*2,
#                cmap = c,ec=ec,
#                alpha=1, lw=0.2) # ec=ec
#     ax.plot(xy[:,0], y_pred, color="#414142", alpha=0.85)
    
    # Add eval metrics as text
    rmse = mean_squared_error(y_true=xy[:,1], y_pred=y_pred, squared=False)
    # rsquared = r2_score(xy[:,1],y_pred)
    pearsonr = scipy.stats.pearsonr(y_pred, xy[:,1])    # Pearson's r
    
    # If first or second plot in top row, add alpha and beta to text
    if plot_idx==0:
        txt = fr"$\beta$ = {np.round(slope,4)}" + "\n" \
        f"{np.round(slope,2)}*X {oper} {np.round(np.absolute(lr_model.intercept_),2)}\n" \
        "$R^2=$ "+ f"{np.round(rsquared,2)}"
        # Add text
        ax.text(.975,.005, txt,
               transform=ax.transAxes, ha="right",va="bottom",
               fontweight="semibold",fontsize=18,
               path_effects=[pe.withStroke(linewidth=6,foreground="white")])
        # return slope
    elif plot_idx==1:
        txt = fr"$\alpha$ = {np.round(slope,2)}" + "\n" \
               f"{np.round(slope,2)}*X {oper} {np.round(np.absolute(lr_model.intercept_),2)}\n" \
               "$R^2=$ " + f"{np.round(rsquared,2)}"
                # Add text
        ax.text(.975,.005, txt,
               transform=ax.transAxes, ha="right",va="bottom",
               fontweight="semibold",fontsize=18,
               path_effects=[pe.withStroke(linewidth=6,foreground="white")])
        alpha = slope
        c_pow = (1 + 2/beta) - alpha
        ax.text(.005,.975, r"(1 + $\frac{2}{\beta}$) - $\alpha$ = " + str(np.round(c_pow,2)),
                   transform=ax.transAxes, ha="left",va="top",
                   fontweight="semibold",fontsize=18,
                   path_effects=[pe.withStroke(linewidth=6,foreground="white")])
        # return slope
    else:
        txt = f"{np.round(slope,2)}*X {oper} {np.round(np.absolute(lr_model.intercept_),2)}\n" \
        "$R^2=$ " + f"{np.round(rsquared,2)}"
        # Add text
        ax.text(.975,.005, txt,
               transform=ax.transAxes, ha="right",va="bottom",
               fontweight="semibold",fontsize=18,
               path_effects=[pe.withStroke(linewidth=6,foreground="white")])

    
#     # Add RLM (HUBER) SKLEARN
#     huber = HuberRegressor(fit_intercept=True, epsilon=1.1).fit(xy[:,0].reshape(-1, 1),xy[:,1])
#     # Predict
#     slope, intercept = huber.coef_[0], huber.intercept_
#     y_pred = slope*xy[:, 0] + intercept
    
#     rsquared = huber.score(xy[:,0].reshape(-1, 1),xy[:,1])
#     # Get RMSE
#     rmse = mean_squared_error(y_true=xy[:,1], y_pred=y_pred, squared=False)
    
#     # Plot 
#     ax.plot(xy[:,0], y_pred, color="#2ca02c", alpha=0.89)

#     txt = f"RLM: y = {np.round(slope,2)} * X + {np.round(intercept,2)}\n" \
#     f"RMSE={np.round(rmse,2)}\n" + r"$R^{2}$=" + str(np.round(rsquared,2))
#     ax.text(0.05, 0.92, txt, ha='left', va='top',
#             transform=ax.transAxes,
#             color = "#2ca02c")
    
    return slope_ols, lr_model.intercept_
```

```python
def get_slope_intercept(x,y):
    
    # Add regression params
    # vstack xy for easier boolean
    xy = np.vstack([x,y]).T
    filterBoolx = np.isfinite(xy[:,0])
    xy = xy[filterBoolx]
    filterBooly = np.isfinite(xy[:,1])    
    xy = xy[filterBooly]
    
    ## Linear regression
    # Get predicted y
    lr_model = LinearRegression(fit_intercept=True).fit(xy[:,0].reshape(-1,1), xy[:,1])
    slope, rsquared = lr_model.coef_[0], lr_model.score(xy[:,0].reshape(-1, 1),
                                                  xy[:,1])
    y_pred = slope*xy[:, 0] + lr_model.intercept_
    slope_ols = slope # to return original intercept
    
    return slope_ols, lr_model.intercept_
```

```python
## Loop through domains
h_dict = {"slope":[],"intercept":[]}
rhc_dict = {"slope":[],"intercept":[]}
rvc_dict = {"slope":[],"intercept":[]}
vc_dict = {"slope":[],"intercept":[]}
rhc_h_dict = {"slope":[],"intercept":[]}
domain_list = []


for domain, group in appidv_all_clean.groupby("domainID"):
    print(domain)
    
    ## Rhc ~ D
    log_x = np.log(group['stemDiameter'])
    log_y = np.log(group["avgCrownDiameter"]/2)
    if np.all(np.isnan(log_x)) or np.all(np.isnan(log_y)):
        print(f"Not enough data in domain: {domain}")
        continue
    try:
        b1, b0 = get_slope_intercept(log_x,log_y)
    except:
        continue
    rhc_dict['slope'].append(b1)
    rhc_dict['intercept'].append(b0)
    
    ## H ~ D
    log_x = np.log(group['stemDiameter'])
    log_y = np.log(group['height'])
    b1, b0 = get_slope_intercept(log_x,log_y)
    h_dict['slope'].append(b1)
    h_dict['intercept'].append(b0)
    
#     ## Rvc ~ H
#     log_x = np.log(group['height'])
#     log_y = np.log(group["ninetyCrownDiameter"]/2)
#     b1, b0 = get_slope_intercept(log_x,log_y)
#     rvc_dict['slope'].append(b1)
#     rvc_dict['intercept'].append(b0)
    
    ## Vc ~ H
    log_x = np.log(group['height'])
    log_y = np.log(group["crown_vol"])
    b1, b0 = get_slope_intercept(log_x,log_y)
    vc_dict['slope'].append(b1)
    vc_dict['intercept'].append(b0)
    
    ## Rhc ~ H
    log_x = np.log(group['height'])
    log_y = np.log(group["avgCrownDiameter"]/2)
    b1, b0 = get_slope_intercept(log_x,log_y)
    rhc_h_dict['slope'].append(b1)
    rhc_h_dict['intercept'].append(b0)
    
    # add domain to list
    domain_list.append(domain)

    
```
