```python
import os
import geopandas as gpd
from osgeo import gdal
import numpy as np
```

```python
#############################
###############################
# Chave
```

```python
# Read AOP polygons
fn = "/data/shared/src/arojas/NEON/data/raw/spatial/AOP_flightBoxes/AOP_flightboxesAllSites.shp"
aop_flightboxes_gdf = gpd.read_file(fn)

# Set vars
# raster_vars = ["CWD", "E"]
raster_var = "E"

# Set up workspace
outdir = "/data/shared/src/arojas/NEON/data/raster/chave"
raster_outdir = os.path.join(outdir,raster_var)
input_raster_fpath = os.path.join("/data/shared/src/arojas/Data/raster/chave",f"{raster_var}.tif")

for site_id in aop_flightboxes_gdf['siteID'].unique():
    # get total bounds, clip raster
    print(site_id)
    
    # Create output folder with sitename
#     site_outdir = os.path.join(raster_outdir,site_id)
#     if not os.path.isdir(site_outdir):
#         os.mkdir(site_outdir)
        
    # Filter for site and reproject to homolosine
    site_flightbox = aop_flightboxes_gdf[aop_flightboxes_gdf['siteID']==site_id].copy()
    site_bbox = np.round(site_flightbox.total_bounds,3)
    
    # Simple raster clip using gdal
    ds = gdal.Open(input_raster_fpath)
#     ds_out = gdal.Translate(os.path.join(raster_outdir, f'{site_id}_{raster_var}.tif'),
#                         ds, projWin = site_bbox, projWinSRS="EPSG:4326")
    ds = gdal.Warp(os.path.join(raster_outdir, f'{site_id}_{raster_var}.tif'),
                   ds, outputBounds = site_bbox, resampleAlg=gdal.GRA_NearestNeighbour)
    ds = None
    ds_out = None
```

```python
## Worldclim for averaged vars

# Read AOP polygons
fn = "/data/shared/src/arojas/NEON/data/raw/spatial/AOP_flightBoxes/AOP_flightboxesAllSites.shp"
aop_flightboxes_gdf = gpd.read_file(fn)

# Set vars
raster_var = "wc2.1_30s_wind"
# raster_var = "wc2.1_30s_srad"
raster_fn = f"{raster_var}_avg_yr.tif"

# Set up workspace
raster_outdir = "/data/shared/src/arojas/NEON/data/raster/worldclim"
input_raster_fpath = os.path.join("/data/shared/src/arojas/Data/raster/worldclim",
                                  raster_var,raster_fn)

for site_id in aop_flightboxes_gdf['siteID'].unique():
    # get total bounds, clip raster
    print(site_id, end="\r")
    
    # Create output folder with sitename
    site_outdir = os.path.join(raster_outdir, site_id)
    if not os.path.isdir(site_outdir):
        os.mkdir(site_outdir)
        
    # Filter for site and reproject to homolosine
    site_flightbox = aop_flightboxes_gdf[aop_flightboxes_gdf['siteID']==site_id].copy()
    site_bbox = np.round(site_flightbox.total_bounds,3)
    
    # Simple raster clip using gdal
    ds = gdal.Open(input_raster_fpath)
#     ds_out = gdal.Translate(os.path.join(raster_outdir, f'{site_id}_{raster_var}.tif'),
#                         ds, projWin = site_bbox, projWinSRS="EPSG:4326")
    ds = gdal.Warp(os.path.join(site_outdir, f'{site_id}_{raster_fn}'),
                   ds, outputBounds = site_bbox, resampleAlg=gdal.GRA_NearestNeighbour)
    ds = None
    ds_out = None
```
