```python
import laspy
import numpy as np
import geopandas as gpd
# Set the grid cells for analysis
import shapely
import numpy as np
import matplotlib.pyplot as plt
```

```python
las = laspy.read("../data/teak_data/lidarAOI_g.las")
```

```python
binx=np.arange(min(las.x),max(las.x),30)
biny=np.arange(min(las.y),max(las.y),30)

xx, yy = np.meshgrid(binx,biny )
```

```python
las_vstack = np.vstack([np.array(las.x), np.array(las.y)]).T
```

```python
# Plot the grid
plt.plot(xx,yy, marker=".",ls="")
plt.grid(True)
plt.scatter(las_vstack[:,0], las_vstack[:,1],marker="+")
```

```python
from scipy.spatial import KDTree
```

```python
T = KDTree(las_vstack)
```

```python
T.data
```

```python
idx = T.query_ball_point([xx[8,9], yy[8,9]],r=30.)
```

```python
import math
math.dist(las_vstack[idx])
```
